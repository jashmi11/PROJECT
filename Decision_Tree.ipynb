```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
movies = pd.read_csv(r'Dataset.csv')
movies
```

```python
label_encoder = LabelEncoder()
movies['genres'] = label_encoder.fit_transform(movies['genres'])
movies['overview'] = label_encoder.fit_transform(movies['overview'])
```

```python
features = ['genres', 'overview']
target = 'popularity'
```

```python
X_train, X_test, y_train, y_test = train_test_split(movies[features], movies[target], test_size=0.2, random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
def make_movie_prediction(model, user_features):
    prediction = model.predict(user_features)
    return prediction
```

```python
user_features = movies[{
    'genres': ['Drama,Crime'],
    'overview': ['Spanning the years 1945 to 1955']}]
```

```python
user_rating_prediction = make_movie_prediction(model, user_features)
movies['predicted_popularity'] = model.predict(movies[features])
```

```python
recommended_movies = movies.sort_values(by='predicted_popularity', ascending=False).head(5)
print("Top 5 Recommended Movies:")
print(recommended_movies['title'].tolist())
```
