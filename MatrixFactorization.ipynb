```python
pivot_table = df.pivot_table(index='title', columns='genre', values='vote_average').fillna(0)
```

```python
pivot_table
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)
resultant_matrix = SVD.fit_transform(pivot_table)
```

```python
corr_mat = np.corrcoef(resultant_matrix)
```

```python
def recommend_movies(movie_title, corr_mat=corr_mat):
    movie_names = pivot_table.index
    movie_list = list(movie_names)
    movie_idx = movie_list.index(movie_title)
    corr_movie_idx = corr_mat[movie_idx]
    recommend = list(movie_names[(corr_movie_idx<1.0) & (corr_movie_idx>0.95)])
    recommend_movies = recommend[:5]
    return recommend_movies
```

```python
print(recommend_movies('The Godfather'))
```
